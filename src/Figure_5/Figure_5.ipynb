The following script is for reading in the results from 10-fold cross-validation for all Active GCA and Healthy control results
as well as for all Inactive GCA and Healthy control results. A scatter plot is constructed for all results.

```r
library("dplyr")
library("ggplot2")
```

```r
#read in the results for active vs. controls random forest classifier
active_vs_control_rf_results = read.csv("../../analysis/classifying_active_vs_control_v2/top_positive_protein_results.csv")
active_vs_control_rf_results$N_Proteins_included <- factor(active_vs_control_rf_results$N_Proteins_included, levels = c("10", "25", "50", "100", "150", "200", "250", "7289"))
active_vs_control_rf_results$Comparison = "Active GCA and Healthy controls"
active_vs_control_rf_results
```

```r
inactive_vs_control_rf_results = read.csv("../../analysis/classifying_inactive_vs_control_v2/top_positive_protein_results.csv")
inactive_vs_control_rf_results$N_Proteins_included <- factor(inactive_vs_control_rf_results$N_Proteins_included, levels = c("10", "25", "50", "100", "150", "200", "250", "7289"))
inactive_vs_control_rf_results$Comparison = "Inactive GCA and Healthy controls"
inactive_vs_control_rf_results
```

```r
#combine all results
all_gca_and_controls_rf_results = rbind(active_vs_control_rf_results, inactive_vs_control_rf_results)
all_gca_and_controls_rf_results
```

```r
# Create a line plot using ggplot2
bar_graph = ggplot(data=all_gca_and_controls_rf_results, aes(x=N_Proteins_included, y=Accuracy, fill = Comparison)) +
            geom_bar(stat = "identity", position = position_dodge(width = 0.9)) + coord_cartesian(ylim = c(0.75, 1)) + 
            geom_text(aes(label = round(Accuracy,3)),position = position_dodge(width = 0.9), vjust = -0.5, angle = 90) 
            
pdf("../../analysis/Figure_5/random_forest_classifier_results.pdf", width = 10, height = 10)
print(bar_graph)
dev.off()
print(bar_graph)
```
